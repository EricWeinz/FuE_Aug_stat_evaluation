```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
org_mean = pd.read_csv("/Users/ericweinzierl/Desktop/Studium/Master/Semester/3.Semester/FuE/Stat_Auswertung/Organ_Excel/mean.csv")
org_std = pd.read_csv("/Users/ericweinzierl/Desktop/Studium/Master/Semester/3.Semester/FuE/Stat_Auswertung/Organ_Excel/std.csv")
org_without_aug = pd.read_csv("/Users/ericweinzierl/Desktop/Studium/Master/Semester/3.Semester/FuE/Stat_Auswertung/Organ_Excel/ohneaug.csv")
org_with_aug = pd.read_csv("/Users/ericweinzierl/Desktop/Studium/Master/Semester/3.Semester/FuE/Stat_Auswertung/Organ_Excel/mitaug.csv")
```

```python
org_mean = org_mean.drop(columns=["test_loss", "test_loss","Next training size","amount original images labelled","amount augmented images labelled","train_loss"])
org_std = org_std.drop(columns=["test_loss", "test_loss","Next training size","amount original images labelled","amount augmented images labelled","train_loss"])
org_without_aug = org_without_aug.drop(columns=["test_loss", "test_loss","Next training size","amount original images labelled","amount augmented images labelled","train_loss"])
org_with_aug = org_with_aug.drop(columns=["test_loss", "test_loss","Next training size","amount original images labelled","amount augmented images labelled","train_loss"])
```

```python
plt.figure(figsize=(8,6))
plt.plot(org_without_aug.epoch, org_without_aug.test_acc, label='Ohne Aug')
plt.plot(org_with_aug.epoch, org_with_aug.test_acc, label='Mit Aug')
plt.plot(org_mean.epoch, org_mean.test_acc, label='Mean')
plt.plot(org_std.epoch, org_std.test_acc, label='Standardabweichung')

plt.legend()
plt.xlabel('Epoch', fontsize=17)
plt.ylabel('Test Accuracy', fontsize=17)
#plt.title("OrganMnist", fontsize=17)
```
