```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mypickle = pd.read_pickle("uncertainty_epoch80_labelled21300.pkl")
mypickle
```

```python
mypickle = pd.read_pickle("uncertainty_epoch=0_labelled=3150.pkl")
mypickle
```

```python
uncertainty = mypickle['uncertainty']
oracle_indices = mypickle['oracle_indices']
labelled_map = mypickle['labelled_map']
```

```python
oracle_indices, uncertainty = (list(t) for t in zip(*sorted(zip(oracle_indices, uncertainty))))
```

```python
print(len(uncertainty))
uncertainty
```

```python
original = uncertainty[0:50000-1]
aug1 = uncertainty[50000:100000-1]
aug2 = uncertainty[100000:150000-1]
```

```python
matrix = np.vstack([original,aug1,aug2])
#matrix     = np.transpose(matrix_aux)
df_lab_img = pd.DataFrame(matrix)
df_lab_img
```

```python
df_lab_img = pd.DataFrame(np.vstack([matrix, df_lab_img.std()]))
df_lab_img
```

```python
df = df_lab_img.transpose()
df.columns = ['original', 'aug1', 'aug2', 'standardabweichung']
df
```

```python
#uncertainties_std.to_excel("uncertainties_std.xlsx")
```

```python
#Uncertainties

uncertainty = pd.DataFrame(uncertainty)
uncertainty.columns = ['uncertainty']
x = uncertainty.index[::100]
y = uncertainty.uncertainty[::100]


plt.figure(figsize=(10,10))
plt.xlabel('Anzahl', fontsize=13)
plt.ylabel('Uncertainties', fontsize=13)
#plt.title("Median der Uncertainties nach Original und Augmentations", fontsize=13)
plt.scatter(x, y, c='crimson', s=20)

z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"b")

```

```python
#Standardabweichungen zw Org und Augs der Uncertainties

index = df.index[::100]
std = df.standardabweichung[::100]

plt.figure(figsize=(10,10))
plt.xlabel('Anzahl', fontsize=13)
plt.ylabel('Uncertainties', fontsize=13)
plt.title("Standardabweichungen der Uncertainties zwischen Original und Augmentations", fontsize=13)

plt.scatter(index, std, c='crimson', s=20)

z = np.polyfit(index, std, 1)
p = np.poly1d(z)
plt.plot(index,p(index),"b")
```

```python
#Median der Uncertainties nach Original und Augmentations

index = df.index[::100]
original = df.original[::100].median()
aug1 = df.aug1[::100].median()
aug2 = df.aug2[::100].median()

#plt.xlabel('Uncertainties', fontsize=17)
plt.ylabel('Uncertainties', fontsize=13)
plt.title("Median der Uncertainties nach Original und Augmentations", fontsize=13)

x = np.array(["Org","Aug1","Aug2"])
y = np.array([original,aug1,aug2])

plt.bar(x, y, width = 0.4)


#plt.scatter(index, original, c='crimson', s=10)
#plt.scatter(index, aug1, c='green',s=10)
#plt.scatter(index, aug2, c='blue',s=10)
```
